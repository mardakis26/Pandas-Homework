### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(20)
```

## Player Count

* Display the total number of players


```python
uniqueplayers = purchase_data["SN"].nunique()
totalplayers = pd.DataFrame({"Total Players":[uniqueplayers]})
totalplayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
sum_unique_items = purchase_data["Item ID"].nunique()
avg_price = purchase_data["Price"].sum()/purchase_data["Price"].count()
num_purchases = purchase_data["Price"].count()
total_revenue = purchase_data["Price"].sum()

purchase_analysis_df = pd.DataFrame({
    "Number of Unique Items":[sum_unique_items],
    "Average Price":[avg_price],
    "Number of Purchases":[num_purchases],
    "Total Revenue":[total_revenue]
})

purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#uniqueplayers is total number of unique players, use for total count
#total male count
totalmales = purchase_data[purchase_data["Gender"]=="Male"]["SN"].nunique()
#male percentage
malepct = (totalmales/uniqueplayers)*100
#total female count
totalfemales = purchase_data[purchase_data["Gender"]=="Female"]["SN"].nunique()
#female percentage
femalepct = (totalfemales/uniqueplayers)*100
#total other/non-disclosed count
totalother = purchase_data[purchase_data["Gender"]=="Other / Non-Disclosed"]["SN"].nunique()
#other/non-disclosed percentage
otherpct = (totalother/uniqueplayers)*100

gender_demographics = pd.DataFrame({"Gender":["Male", "Female", "Other / Non-Disclosed"],
                                   "Total Count": [totalmales, totalfemales, totalother],
                                    "Percentage of Players":[malepct, femalepct, otherpct]
                                   })
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#number of purchases per gender
genderpurchases = purchase_data.groupby("Gender")["Item Name"].count()

#avg purchase price per gender
genderpricepurch = purchase_data.groupby("Gender")["Price"].mean()

#total purchase value per gender
genderpurchvalue = purchase_data.groupby("Gender")["Price"].sum()

#purchase total per person, per gender
purchpp = genderpurchvalue/len(totalplayers)

#display data in cleaner way

gender_purchasing_analysis = pd.DataFrame({
    "Number of Purchases per Gender":genderpurchases,
    "Average Purchase Price per Gender":genderpricepurch,
    "Total Purchase Value per Gender":genderpurchvalue,
    "Purchase Total per Person per Gender":purchpp
})

gender_purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#establish bins for ages
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
labels = ["<10", "10-15", "16-20", "21-25", "26-30", "31-35", "36-40", "41+"]

pd.cut(purchase_data["Age"], bins, labels=labels)

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=labels)
purchase_data.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
agerangedata = purchase_data.groupby("Age Range")

#purchase count by age range
agepurchcount = purchase_data.groupby("Age Range")["Item ID"].nunique()

#average purchase price by age range
ageavgprice = purchase_data.groupby("Age Range")["Price"].mean()

#average purchase total per person per age
agepurch = purchase_data.groupby("Age Range")["Price"].sum()
agepurchpp = agepurch/agepurchcount

age_purchasing_analysis = pd.DataFrame ({
    "Number of Purchases":(agepurchcount),
    "Average Purchase Price":(ageavgprice),
    "Total Purchase Value":(agepurch),
    "Average Purchase Total per Person":(agepurchpp)
})


age_purchasing_analysis.head()

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
players_purchase = purchase_data.groupby("SN")

#number of purchases per player
players_purchase_count = players_purchase["Price"].count()
#total purchase value per player
players_purchase_total = players_purchase["Price"].sum()
#average purchase price per player
players_purchase_avg = players_purchase["Price"].mean()

top_spenders = pd.DataFrame ({
    "Number of Purchases":(players_purchase_count),
    "Average Purchase Price":(players_purchase_avg),
    "Total Purchase Value":(players_purchase_total)
})

top_spenders.head()

top_spenders.sort_values(by="Total Purchase Value", ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



## Most Profitable Items

```python
#Most Popular Items

#Retrieve the Item ID, Item Name, and Item Price columns
items = purchase_data[["Item ID", "Item Name", "Price"]]

#purchase count
items_count = items.groupby(["Item ID", "Item Name"]).count()["Price"]

#item price
items_price = items.groupby(["Item ID", "Item Name"]).mean()["Price"]

#total purchase value
items_total = items.groupby(["Item ID", "Item Name"]).sum()["Price"]

popular_items = pd.DataFrame({
    "Price":(items_price),
    "Purchase Count":(items_count),
    "Total Purchase Value":(items_total)
})

popular_items.head()
popular_items.sort_values(by="Purchase Count", ascending=False)
```
